## 2. What is Driver Drowsiness Detection (Technical Summary)

It is a **computer vision and AI-based safety system** that monitors:

* Eye closure (Eye Aspect Ratio – EAR)
* Blink duration
* Yawning frequency (Mouth Aspect Ratio – MAR)

Using:

* Camera input
* Facial landmark detection (MediaPipe / Dlib / OpenCV)
* Machine Learning or Deep Learning models

The system triggers:

* **Audio alerts**
* **Seat vibration**
* **Dashboard warnings**

```python
import cv2
import mediapipe as mp
import math
import winsound
```

```python
def EYE_ASPECT(points):
    A=math.dist(points[1],points[5])
    B=math.dist(points[2],points[4])
    C=math.dist(points[0],points[3])
    return (A+B)/(2*C)
```

```python
def MOUTH_ASPECT(points):
    A=math.dist(points[1],points[7])
    B=math.dist(points[2],points[6])
    C=math.dist(points[3],points[5])
    D=math.dist(points[0],points[4])
    return (A+B+C)/(2*D)
```

```python
#Load the model
facemesh=mp.solutions.face_mesh.FaceMesh(static_image_mode=False,
    max_num_faces=1,
    refine_landmarks=False,
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5)
face_detection=mp.solutions.face_detection.FaceDetection(min_detection_confidence=0.5,model_selection=1)
```

```python
#Input to the model WEB CAMERA
video=cv2.VideoCapture(0)
left_eye=[33,160,158,133,153,144]
right_eye=[362,385,387,263,373,380]
mouth=[61,81,13,311,291,308,402,14]
eye_blink_counter=0
yawn_count=0
while video.isOpened():
    success,frame=video.read()
    frame=cv2.resize(frame,(640,480))
    if not success:
        break
    rgb_image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    result=facemesh.process(rgb_image)
    results=face_detection.process(rgb_image)
    if results.detections:
        for detection in results.detections:
            bbox=detection.location_data.relative_bounding_box
            xmin=int(bbox.xmin*frame.shape[1])
            ymin=int(bbox.ymin*frame.shape[0])
            xmax=int(bbox.width*frame.shape[1])
            ymax=int(bbox.height*frame.shape[0])
            cv2.rectangle(frame,pt1=(xmin,ymin),pt2=(xmin+xmax,ymin+ymax),color=(0,255,0),thickness=2)
    if result.multi_face_landmarks:
        for detection in result.multi_face_landmarks:
            h=frame.shape[0]
            w=frame.shape[1]
            landmarks=[(int(i.x*w),int(i.y*h)) for i in detection.landmark]
            #print(landmarks)
            left=[landmarks[i] for i in left_eye]
            right=[landmarks[i] for i in right_eye]
            mouthh=[landmarks[i] for i in mouth]
            left_eyee=EYE_ASPECT(left)
            right_eyee=EYE_ASPECT(right)
            moutth=MOUTH_ASPECT(mouthh)
            avg_eye=(left_eyee+right_eyee)/2
            if avg_eye<0.25:
                eye_blink_counter=eye_blink_counter+1
            else:
                eye_blink_counter=0
            if moutth>0.6:
                yawn_count=yawn_count+1
                cv2.putText(frame,"Driver is Yawning",org=(30,80),fontFace=cv2.FONT_HERSHEY_DUPLEX,fontScale=1,color=(0,255,0),thickness=2)
            else:
                yawn_count=0
            if(eye_blink_counter > 30) or (yawn_count > 15):
                winsound.Beep(2000,500)
                cv2.putText(frame,"Driver is Drowsy",org=(30,80),fontFace=cv2.FONT_HERSHEY_DUPLEX,
                    fontScale=1,color=(0,255,0),thickness=2)
    cv2.imshow("WEB CAM",frame)
    if cv2.waitKey(1) & 255 == ord("q"):
        break
video.release()
cv2.destroyAllWindows()
```
